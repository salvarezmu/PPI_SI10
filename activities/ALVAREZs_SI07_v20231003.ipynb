## gráfico 00

Seleccione dos columnas numéricas del dataset y haga un gráfico tipo **plot**. Coloque título al gráfico y etiquete los dos ejes.

```
import matplotlib.pyplot as plt


def build_grafico00(mpl_element, set_size=False, is_axe=True):
    """La función genera un gráfico tipo plot con dos columnas\
    númericas.

    El parámetro set_size indica si se debe cambiar el tamaño de la figura.
    El parámetro is_axe indica si se está ejecutando desde un subplot.
    """

    # Se aumenta el tamaño de la figura
    if set_size:
        mpl_element.figure(figsize=(10, 6))

    # Se genera el gráfico 00 de tipo plot con la salud\
    # y defensa de los pokemon
    mpl_element.plot(pokemon_dataset.loc[:, ['Health Stat', 'Defense Stat']])

    # Se establece la etiqueta de los ejes y el título del gráfico
    if is_axe:
        mpl_element.set_xlabel('Salud del Pokemon')
        mpl_element.set_ylabel('Defensa del Pokemon')
        mpl_element.set_title('Pokemon - Salud vs Defensa')

    else:
        mpl_element.xlabel('Salud del Pokemon')
        mpl_element.ylabel('Defensa del Pokemon')
        mpl_element.title('Pokemon - Salud vs Defensa')


# Se genera el gráfico requerido
build_grafico00(plt, True, False)

# Se muestra el gráfico
plt.show()
```

## gráfico 01

Seleccione cinco columnas numéricas del dataset y haga un gráfico tipo **scatter**. Coloque título al gráfico y etiquete los dos ejes.

```
import matplotlib.pyplot as plt

# Se calculan los promedios de las columnas escogídas
g01_values = [
    pokemon_dataset['Health Stat'].mean(),
    pokemon_dataset['Attack Stat'].mean(),
    pokemon_dataset['Special Attack Stat'].mean(),
    pokemon_dataset['Special Defense Stat'].mean(),
    pokemon_dataset['Speed Stat'].mean()
]

# Se seleccionan los títulos de las columnas
g01_titles = [
    'Salud', 'Ataque', 'Ataque especial',
    'Defensa especial', 'velocidad'
    ]


def build_grafico01(mpl_element, set_size=False, is_axe=True):
    """La función genera un gráfico tipo scatter con cinco columnas\
    númericas.

    El parámetro set_size indica si se debe cambiar el tamaño de la figura.
    El parámetro is_axe indica si se está ejecutando desde un subplot.
    """

    # Se aumenta el tamaño de la figura
    if set_size:
        mpl_element.figure(figsize=(10, 6))

    # Se genera el gráfico 01 de tipo scatter
    grafico01 = mpl_element.scatter(g01_titles, g01_values, s=g01_values)

    # Se establece la etiqueta de los ejes y el título del gráfico
    if is_axe:
        mpl_element.set_xlabel('Estadísticas')
        mpl_element.set_ylabel('Promedios')
        mpl_element.set_title('Media de estadísticas Pokemon principales')

    else:
        mpl_element.xlabel('Estadísticas')
        mpl_element.ylabel('Promedios')
        mpl_element.title('Media de estadísticas Pokemon principales')


# Se genera el gráfico requerido
build_grafico01(plt, True, False)

# Se muestra el gráfico
plt.show()
```

## gráfico 02

Seleccione cuatro columnas numéricas del dataset y haga un gráfico tipo **stackplot** (apilable). Coloque título al gráfico, etiquete los dos ejes e incluya una leyenda dentro del gráfico.

```
import matplotlib.pyplot as plt

# Se genera la cantidad de datos en el eje X
x = list(range(0, 1235))

# Se seleccionan las columnas necesarias
g02_healt = pokemon_dataset['Health Stat']
g02_attack = pokemon_dataset['Attack Stat']
g02_special_attack = pokemon_dataset['Special Attack Stat']
g02_special_defense = pokemon_dataset['Special Defense Stat']

# Se sekeccuibab los nombres de las columnas
g02_titles = ['Salud', 'Ataque', 'Ataque especial', 'Defensa especial']


def build_grafico02(mpl_element, set_size=False, is_axe=True):
    """La función genera un gráfico tipo stackplot con cuatro columnas\
    númericas.

    El parámetro set_size indica si se debe cambiar el tamaño de la figura.
    El parámetro is_axe indica si se está ejecutando desde un subplot.
    """

    # Se aumenta el tamaño de la figura
    if set_size:
        mpl_element.figure(figsize=(20, 9))

    # Se crea la figura
    grafico02 = mpl_element.stackplot(
        x,
        g02_healt,
        g02_attack,
        g02_special_attack,
        g02_special_defense,
        labels=g02_titles
        )
    mpl_element.legend(loc='upper left')

    # Se establece la etiqueta de los ejes y el título del gráfico
    if is_axe:
        mpl_element.set_xlabel('Estadísticas')
        mpl_element.set_ylabel('Valor')
        mpl_element.set_title('Stack de estadísticas principales')

    else:
        mpl_element.xlabel('Índice del Pokemon')
        mpl_element.ylabel('Valor')
        mpl_element.title('Stack de estadísticas principales')


# Se genera el gráfico requerido
build_grafico02(plt, True, False)

# Se muestra el gráfico
plt.show()
```

## gráfico 04

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico de barras horizontales. Coloque título al gráfico y etiquete los dos ejes.

```
import matplotlib.pyplot as plt

# Se define el ancho de las barras
g04_bar_width = 0.40

# Se obtienen los datos para el gráfico
g04_primary = pokemon_dataset.groupby('Primary Type')['Primary Type']
g04_secondary = pokemon_dataset.groupby('Secondary Type')['Secondary Type']

# Se obtienen las categorías del gráfico
g04_categories = g04_primary.agg('count').index.to_list()

# Se aumenta el tamaño de la figura
plt.figure(figsize=(10, 9))

# Se generan las barras de la columna Primary Type
plt.barh(
    g04_categories,
    g04_primary.agg('count'),
    g04_bar_width,
    label='Tipo Primario'
    )

# Se generan las barras de la columna Secondary Type
g04_secondoray_sizes = [i + g04_bar_width for i in range(len(g04_categories))]
plt.barh(
    g04_secondoray_sizes,
    g04_secondary.agg('count'),
    g04_bar_width,
    label='Tipo Secundario'
    )

# Se muestran los labels de cada barra
plt.legend(loc='lower right')

# Se establece la etiqueta de los ejes y el título del gráfico
plt.xlabel('Frecuencia')
plt.ylabel('Tipo Pokemon')
plt.title('Frecuencia del tipo Pokemon primario y secundario')

# Se muestra el gráfico
plt.show()
```

## gráfico 07

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico de torta (**pie chart**). Coloque título al gráfico y etiquete los segmentos del gráfico.

```
import matplotlib.pyplot as plt

# Se obtiene el grupo para generar el gráfico
g07_values = pokemon_dataset.groupby('Primary Type')['Primary Type']
g07_values = g07_values.agg('count')

# Se generan los porcentajes para el gráfico de torta
g07_values = g07_values * 100 / pokemon_dataset.shape[0]

# Se obtienen las categorías del gráfico
g07_categories = g07_values.index.to_list()

# Se aumenta el tamaño de la figura
plt.figure(figsize=(12, 12))

# Se genera el gráfico de torta
plt.pie(g07_values, labels=g07_categories, autopct='%1.1f%%', startangle=140)

# Se establece el título del gráfico
plt.title('Tipos Pokemon')

# Se muestra el gráfico
plt.show()
```

## gráfico 08

Seleccione las columnas que considere pertinentes del dataset y haga un mapa de calor (**heatmap**). Coloque título al gráfico y etiquete los dos ejes.

```
import matplotlib.pyplot as plt

# Se obtienen las columnas para generar el gráfico
g08_colums = pokemon_dataset.loc[:, [
    'Health Stat',
    'Attack Stat',
    'Defense Stat',
    'Special Attack Stat',
    'Special Defense Stat',
    'Speed Stat'
    ]]

# Se genera la correlación entre las columnas
g08_data = g08_colums.corr()

# Se genera el mapa de calor
plt.imshow(g08_data, cmap='coolwarm', interpolation='nearest')

# Se establece la etiqueta de los ejes y el título del gráfico
plt.title('Correlación entre estadísticas de \
atributos del Pokemon')
plt.xlabel('Estadísticas de Atributos')
plt.ylabel('Estadísticas de Atributos')

# Se muestra el gráfico
plt.show()
```

## gráfico 09

Seleccione las columnas que considere pertinentes del dataset y haga un gráfico en 3D. Coloque título al gráfico y etiquete los ejes.

```
import matplotlib.pyplot as plt

# Se obtienen las columnas de interés
g09_height = pokemon_dataset['Pokemon Height']
g09_weight = pokemon_dataset['Pokemon Weight']
g09_stat_total = pokemon_dataset['Base Stat Total']

# Se crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Crear el gráfico de burbujas 3D
ax.scatter(
    g09_height,
    g09_weight,
    g09_stat_total,
    c='b',
    marker='o',
    label='Pokémon'
    )

# Se etiquetan los ejes y se establece el título del gráfico
ax.set_xlabel('Altura del Pokémon')
ax.set_ylabel('Peso del Pokémon')
ax.set_zlabel('Total de estadísticas básicas')
ax.set_title('Gráfico de Burbujas 3D: Altura vs.\
 Peso vs. Total de estadísticas básicas')

# Se muestra el gráfico
plt.show()
```
